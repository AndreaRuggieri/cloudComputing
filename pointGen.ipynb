```python
import matplotlib.pyplot as plt

def read_points_from_file(file_name):
    points = []
    with open(file_name, 'r') as file:
        for line in file:
            values = line.strip().split(', ')
            x = float(values[0])
            y = float(values[1])
            points.append((x, y))
    return points

def plot_points(points):
    x = [point[0] for point in points]
    y = [point[1] for point in points]
    plt.scatter(x, y, s=5)  # Adjust the 's' parameter to change the size of the points
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.title('Points')
    plt.show()

# Provide the path to your text file
file_path = './input.txt'

points = read_points_from_file(file_path)
plot_points(points)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Read objects from file
objects = []
with open('./input.txt', 'r') as file:
    for line in file:
        values = line.strip().split(',')
        objects.append([float(values[0]), float(values[1]), float(values[2])])

# Extract x, y, and z coordinates
x_coords = [obj[0] for obj in objects]
y_coords = [obj[1] for obj in objects]
z_coords = [obj[2] for obj in objects]

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the objects
ax.scatter(x_coords, y_coords, z_coords)

# Set labels and title
ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_zlabel('Z Coordinate')
ax.set_title('Objects in 3D Cartesian Graph')

# Show the plot
plt.show()
```
